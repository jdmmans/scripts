```python
g =  [arr[0],arr[2]]
k = [arr[1],arr[3]]
```

```python
f_array = np.column_stack((g,k))
f_array
```

```python
g1 = np.asarray(g)
k1 = np.asarray(k)
```

```python
np.column_stack((g1,k1))
```

```python
 import scipy.stats as stats
```

```python
# fishers test (odds ratio, p value)
oddsratio, pvalue = stats.fisher_exact(f_array)
pvalue
```

```python
stats.fisher_exact([[2,7],[31,98]])
```

```python
# chi squared (chi squared, p-value, degrees of freedom, expected array frequencies)
from scipy.stats import chi2_contingency
chi2_contingency(f_array)
```

```python
chi2_contingency([[2,7,9],[20,70,0]])
```
